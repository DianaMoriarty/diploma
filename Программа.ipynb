# Программа конференций

Необходимые модули

```python
import psycopg2
import pandas as pd
```

Функция для подключения к базе данных и выполнения запроса.

```python
def connection_to_db(query):
    conn = psycopg2.connect(host="localhost", port = 5432, database="postgres", user="postgres", password="sql123")
    cur = conn.cursor()
    cur.execute(query)
    query_results = cur.fetchall()
    cur.close()
    conn.close()
    return query_results
```

Функция для выборки по дням.

```python
def program_by_days(day):
    query = "SELECT to_char(start_time, 'HH24:MI') as time, author FROM diploma.program WHERE to_char(start_time,'DD.MM.YYYY') = %s ORDER BY time;" % day
    program = connection_to_db(query)
    return program
```

Функция для выборки по секциям

```python
def program_by_section(section):
    query = "SELECT to_char(start_time, 'HH24:MI') as time, to_char(start_time, 'DD.MM.YYYY'), author FROM diploma.program WHERE section_id = %s ORDER BY start_time;" % section
    program = connection_to_db(query)
    return program
```

Функция для выборки по залам

```python
def program_by_hall(hall):
    query = "SELECT to_char(start_time, 'HH24:MI') as time, to_char(start_time, 'DD.MM.YYYY'), author FROM diploma.program WHERE hall_id = %s ORDER BY start_time;" % hall
    program = connection_to_db(query)
    return program
```

# Примеры

```python
day = "'14.09.2020'"
data = pd.DataFrame(program_by_days(day), columns = ['Время', 'Автор'])
```
